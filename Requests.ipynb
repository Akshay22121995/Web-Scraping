```python
import requests
```

```python
r = requests.get('https://quotes.toscrape.com')
```

```python
print(r.status_code) #It prints the response code of web page
```

```python
print(r.text) # Prints the Content of web page
```

- When we fetch the web content, we get complete response as string
- Now we have to split that complete string into seperate lines using  '\n'

```python
html = r.text
```

```python
for line in html.split('\n'):
    if '<span class="text" itemprop="text">' in line:
        line = line.replace('<span class="text" itemprop="text">','').replace('</span>','') #Here we replace the starting and the ending tag
        line = line.strip() # It will remove all the whitespaces
        print(line)
```

### Now we are going to save these quotes in a file

```python
with open('quotes.txt','w') as f:
    for line in html.split('\n'):
        if '<span class="text" itemprop="text">' in line:
            line = line.replace('<span class="text" itemprop="text">“','').replace('”</span>','') #Here we replace the starting and the ending tag
            line = line.strip() # It will remove all the whitespaces
            f.write(line)
            f.write('\n')
```

### Extracting Authors Name

```python
for line in html.split('\n'):
        if '<small class="author" itemprop="author">' in line:
            line = line.replace('<span>by <small class="author" itemprop="author">','').replace('</small>','') #Here we replace the starting and the ending tag
            authors = line.strip()
            print(authors)
```

### Save Author Names in file

```python
with open('author_names.txt','w')as f:
    for line in html.split('\n'):
        if '<small class="author" itemprop="author">' in line:
            line = line.replace('<span>by <small class="author" itemprop="author">','').replace('</small>','') #Here we replace the starting and the ending tag
            authors = line.strip()
            f.write(authors)
            f.write('\n')
```

### We are Extracting the data from multiple pages by Extracting the changing URL 
> If we want to fetch too much web pages then Request is very slow because it executes sequentially (i.e. It sends the request to get the URL and waits for the server to respond and then it will send the next web page URL)

```python
with open('All quotes.txt','w') as f:
    for i in range(10):
        url = f'https://quotes.toscrape.com/page/{i}/'
        links = requests.get(url)
        html = link.text
        for line in html.split('\n'):
            if '<span class="text" itemprop="text">' in line:
                line = line.replace('<span class="text" itemprop="text">“','').replace('”</span>','') #Here we replace the starting and the ending tag
                line = line.strip() # It will remove all the whitespaces
                f.write(line)
                f.write('\n')
```

### Extracting Author names and their Quotes

```python
with open('Author and Quotes.csv','w') as f:
    for i in range(10):
        url = f'https://quotes.toscrape.com/page/{i}/'
        links = requests.get(url)
        html = link.text
        for line in html.split('\n'):
            if '<span class="text" itemprop="text">' in line:
                    line = line.replace('<span class="text" itemprop="text">“','').replace('”</span>','') #Here we replace the starting and the ending tag
                    quotes = line.strip()
            if '<small class="author" itemprop="author">' in line:
                    line = line.replace('<span>by <small class="author" itemprop="author">','').replace('</small>','') #Here we replace the starting and the ending tag
                    authors = line.strip()
                    #print(authors,' : ',quotes)
                    f.write(authors)
                    f.write(',')
                    f.write(quotes)
                    f.write('\n')
                
```
